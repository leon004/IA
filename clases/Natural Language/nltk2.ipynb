```python
import nltk
import twint
import nest_asyncio
nest_asyncio.apply()
```

```python
c = twint.Config()
c.Search = "PelonGomis"
c.Limit = 100
c.Pandas = True

twint.run.Search(c)
```

```python
df = twint.storage.panda.Tweets_df
tweet = df['tweet']
tweet = tweet.to_list()
text = ''
for i in tweet:
    text = text + i
```

```python
tokens = nltk.word_tokenize(text)
text2 = nltk.Text(tokens)
text2
```

```python
text2.concordance('chayoteros')
```

```python
text2.similar('chayotero')
```

```python
text2.common_contexts(['VicenteSerrano'])
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import sentiment
from nltk import word_tokenize
```

```python
analizador = SentimentIntensityAnalyzer()
```

```python
for twit in text2:
    print(twit)
    scores = analizador.polarity_scores(twit)
    for key in scores:
        print(key, ':', scores[key])
        print()
    
```

## ¬øQu√© es NLTK Sentiment Vader?
VADER (Valence Aware Dictionary for Sentiment Reasoning) es un modelo utilizado para el an√°lisis de sentimientos de texto que es sensible tanto a la polaridad (positiva / negativa) como a la intensidad (fuerza) de la emoci√≥n. Est√° disponible en el paquete NLTK y se puede aplicar directamente a datos de texto sin etiquetar.

VADER(Valence Aware Dictionary and sEntiment Reasoner) es una herramienta de an√°lisis de sentimientos basada en reglas y l√©xico que se adapta espec√≠ficamente a los sentimientos expresados en las redes sociales. VADER usa una combinaci√≥n de Un l√©xico de sentimiento es una lista de caracter√≠sticas l√©xicas(por ejemplo, palabras) que generalmente se etiquetan de acuerdo con su orientaci√≥n sem√°ntica como positiva o negativa.
```python
c = twint.Config()
c.Search = "Trump"
#c.Lang = "en"
c.Limit = 10
c.Pandas = True

twint.run.Search(c)
```

```python
df = twint.storage.panda.Tweets_df
tweet = df['tweet']
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import sentiment
from nltk import word_tokenize
```

```python

for twit in tweet:
    scores = analizador.polarity_scores(twit)
    print("{} - score: {}".format(twit, scores))
```
