```python
import os

# Cambiar al directorio indicado
os.chdir("/content/gdrive/MyDrive")

# Lista de carpetas a crear dentro de "/content/gdrive/MyDrive"
folders_to_create = [
    "FPRDataset",
    "FPRDataset/train",
    "FPRDataset/train/nodule",
    "FPRDataset/train/non-nodule",
    "FPRDataset/test",
    "FPRDataset/test/nodule",
    "FPRDataset/test/non-nodule"
]

# Crear las carpetas si no existen
for folder in folders_to_create:
    # Construir el path completo para la carpeta
    folder_path = os.path.join("/content/gdrive/MyDrive", folder)

    if not os.path.exists(folder_path):
        # Si la carpeta no existe, crearla (incluyendo todos los directorios intermedios necesarios)
        os.makedirs(folder_path)
        print(f"La carpeta {folder_path} ha sido creada.")
    else:
        print(f"La carpeta {folder_path} ya existe.")
```

```python
import os

# Rutas para las carpetas 'nodule' y 'non-nodule'
nodule_path = "/content/gdrive/MyDrive/FPRProcessed/nodule/"
non_nodule_path = "/content/gdrive/MyDrive/FPRProcessed/non-nodule/"

# Crear la carpeta 'nodule' si no existe
if not os.path.exists(nodule_path):
    os.makedirs(nodule_path)
    print(f"La carpeta {nodule_path} ha sido creada.")

# Crear la carpeta 'non-nodule' si no existe
if not os.path.exists(non_nodule_path):
    os.makedirs(non_nodule_path)
    print(f"La carpeta {non_nodule_path} ha sido creada.")


nodule_files = os.listdir(nodule_path)
non_nodule_files = os.listdir(non_nodule_path)
```

```python
nodule_files = os.listdir("/content/gdrive/MyDrive/FPRProcessed/nodule/")
non_nodule_files = os.listdir("/content/gdrive/MyDrive/FPRProcessed/non-nodule/")
```

```python
random.shuffle(nodule_files)
random.shuffle(non_nodule_files)
```

```python
n = len(nodule_files)
split = int(0.75*n)
for i in nodule_files[:split]:
    shutil.copy(f"/content/gdrive/MyDrive/FPRProcessed/nodule/{i}", f"/content/gdrive/MyDrive/FPRDataset/train/nodule/{i}")
for i in nodule_files[split:]:
    shutil.copy(f"/content/gdrive/MyDrive/FPRProcessed/nodule/{i}", f"/content/gdrive/MyDrive/FPRDataset/test/nodule/{i}")
```

```python
n = len(non_nodule_files)
split = int(0.75*n)
for i in non_nodule_files[:split]:
    shutil.copy(f"/content/gdrive/MyDrive/FPRProcessed/non-nodule/{i}", f"/content/gdrive/MyDrive/FPRDataset/train/non-nodule/{i}")
for i in non_nodule_files[split:]:
    shutil.copy(f"/content/gdrive/MyDrive/FPRProcessed/non-nodule/{i}", f"/content/gdrive/MyDrive/FPRDataset/test/non-nodule/{i}")
```
